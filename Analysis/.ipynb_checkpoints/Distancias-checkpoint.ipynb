```python
import pandas as pd
from math import sin, cos, sqrt, atan2, radians
import numpy as np
from sklearn.datasets import load_digits
from sklearn.manifold import MDS
from scipy.stats.stats import pearsonr
import matplotlib.pyplot as plt
import math
import utm 
from skbio.stats.distance import permanova
from skbio import DistanceMatrix

%matplotlib inline

def gen_distance(mdss, principalDf2, sites, dissimilarity):
    mdss=["MDS component 1"]
    fake_distance=[]
    real_distance=[]
    for pin in range(len(sites)):
        for uin in range(pin+1, len(sites)):
            p=sites[pin]
            u=sites[uin]
            distance=0.0
            for k in mdss:
                distance+=(principalDf2.loc[p, k]-principalDf2.loc[u, k])**2
            distance=distance**0.5
            fake_distance.append(distance)
            real_distance.append(dissimilarity[pin][uin])
    return real_distance, fake_distance

def addplot(tupleloc, col1, col2, principalDf2, title="", subtitle="", spcol=False, labels=False):
    if title==False: title="Unkown matrix"
    ax = fig.add_subplot(*tupleloc) 
    ax.set_xlabel(col1, fontsize = 10)
    ax.set_ylabel(col2, fontsize = 10)
    ax.set_title(' '+title+'\n '+subtitle, fontsize = 10)
    if spcol==False:
        ax.scatter(principalDf2.loc[:,col1]
                   , principalDf2.loc[:, col2]
                   , c = 'r'
                   , s = 50 )
    else:
        vals=set(principalDf2[spcol])
        colores=['r', 'b', 'g']
        inc=0
        for val in vals:
            ax.scatter(principalDf2.loc[principalDf2[principalDf2[spcol]==val].index,col1]
                   , principalDf2.loc[principalDf2[principalDf2[spcol]==val].index, col2]
                   , c = colores[inc]
                   , s = 50 , label=labels[val])
            ax.legend(bbox_to_anchor=(1, -0.1))
            inc+=1
        

    for i in principalDf2.iterrows():
        ax.annotate(i[0], (i[1][col1], i[1][col2]))

    ax.grid()
    
    
def addplot_sp(tupleloc, v1, v2, col1, col2, title="", subtitle=""):
    if title==False: title="Unkown matrix"
    ax = fig.add_subplot(*tupleloc) 
    ax.set_xlabel(col1, fontsize = 10)
    ax.set_ylabel(col2, fontsize = 10)
    ax.set_title(' '+title+'\n '+subtitle, fontsize = 10)
    ax.scatter(v1
               , v2
               , c = 'r'
               , s = 50 )


    ax.grid()
    
def genpcdf(n_components, indexes, other_information,  precomputed=False, euclidean=False, max_iter=3000, n_init=50, return_embedding=False):
    if precomputed==False:
        embedding = MDS(n_components=n_components,max_iter=max_iter, n_init=n_init , random_state=5676 , metric=False)
        X_transformed = embedding.fit_transform(euclidean)
        
    else:
        embedding = MDS(n_components=n_components, max_iter=3000, n_init=50, random_state=5676 , metric=False, dissimilarity="precomputed")
        X_transformed = embedding.fit_transform(precomputed)
        
    principalDf2 = pd.DataFrame(data = X_transformed
                 , columns = ['MDS component '+str(ind) for ind in range(1,n_components+1)], index=indexes)

    principalDf2=principalDf2.join(other_information)
    real_distance, fake_distance=gen_distance(mdss, principalDf2, indexes, embedding.dissimilarity_matrix_)

    return {"df":principalDf2, "stress":embedding.stress_, "real_distance":real_distance, "simulated_distance":fake_distance}

def get_euclidean(matrix):
    embedding = MDS(n_components=2, max_iter=3000, n_init=50, random_state=5676 , metric=False)
    X_transformed = embedding.fit_transform(matrix)
    disimilairity_euclidean=embedding.dissimilarity_matrix_
    return disimilairity_euclidean

coordenadas={"Marrickville":(-33.90468, 151.16067),
            "Surry Hills":(-33.88972, 151.20965),
            "Camperdown":(-33.88514, 151.18837), 
            "Asheville":(-33.89428, 151.12279),
            "Balgowlah ":(-33.79207, 151.26724),
            "Manley":(-33.80352, 151.28617),
            "Jannali":(-34.01215, 151.07181),
            "Sutherland":(-34.03502, 151.05768), 
            "Stives": (-33.70606, 151.18046),
            "Palmbeach":(-33.66858, 151.31378),
            "Lindfield":(-33.77102, 151.16314),
            "Gordon":(-33.75781, 151.15081),
            "Caringbah":(-34.06095, 151.11111)}

distance_national_park={
    "Marrickville":17.5,
            "Surry Hills":20,
            "Camperdown":20, 
            "Asheville":17.5,
            "Balgowlah ":11.3,
            "Manley":13.5,
            "Jannali":3.3,
            "Sutherland":5, 
            "Stives": 0.1,
            "Palmbeach":3.4,
            "Lindfield":7.3,
            "Gordon":6.5,
            "Caringbah":1
}


xaxis  = [utm.from_latlon(coordenadas[l][0],coordenadas[l][1])[0]  for l in coordenadas]
yaxis = [utm.from_latlon(coordenadas[l][0],coordenadas[l][1])[1] for l in coordenadas]
medy  =  np.mean(yaxis)
medx  =  np.mean(xaxis)
#xaxis= [(i-med)/1000 for i in xaxis]

#yaxis = [(i-med)/1000 for i in yaxis]


nl=[]
for p in distance_national_park:
    coords=utm.from_latlon(coordenadas[p][0],coordenadas[p][1])
    nd={"Index":p, "Distance_national_park":distance_national_park[p], "Latitud":coordenadas[p][0], "Longitud":coordenadas[p][1], "X":(coords[0]-medx)/1000, "Y":(coords[1]-medy)/1000, "near_np":int(distance_national_park[p]<10), "super_near_np":int(distance_national_park[p]<3)}
    nl.append(nd)
information=pd.DataFrame(nl).set_index("Index")

def to_xy(point):
    r=6371000
    phi_0 = 151.16067
    cos_phi_0 = math.cos(math.radians(phi_0))
    lam = point[0]
    phi = point[1]
    return (r * math.radians(lam) * cos_phi_0, r * math.radians(phi))


def physical_distance(p1, p2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(p1[0])
    lon1 = radians(p1[1])
    lat2 = radians(p2[0])
    lon2 = radians(p2[1])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

abundance_t=pd.read_csv("Tomar_vuelo.csv")
abundance_t=abundance_t.set_index("Familia")
abundance=abundance_t.transpose()
abundance_dict=abundance.to_dict()

for k in abundance_dict:
    for p in abundance_dict[k]:
        abundance_dict[k][p]=int(abundance_dict[k][p]>0)
absence=pd.DataFrame(abundance_dict)

abundance_dict=abundance.to_dict()
for k in abundance_dict:
    for p in abundance_dict[k]:
        abundance_dict[k][p]=int(abundance_dict[k][p]>1)
absence_2=pd.DataFrame(abundance_dict)

abundance_dict=abundance.to_dict()
for k in abundance_dict:
    for p in abundance_dict[k]:
        abundance_dict[k][p]=int(abundance_dict[k][p]>2)
absence_3=pd.DataFrame(abundance_dict)

abundance_dict=abundance.to_dict()
for k in abundance_dict:
    for p in abundance_dict[k]:
        abundance_dict[k][p]=int(abundance_dict[k][p]>3)
absence_4=pd.DataFrame(abundance_dict)


abundance_t_dict=abundance_t.to_dict()
for k in abundance_t_dict:
    tot=np.sum(list(abundance_t_dict[k].values()))
    for p in abundance_t_dict[k]:
        abundance_t_dict[k][p]=abundance_t_dict[k][p]/float(tot)
normalized_abundance=pd.DataFrame(abundance_t_dict).transpose()

sites=list(abundance.index)

```

```python
def exponential_distribution(lam, x):
    return lam*math.exp(-lam*x)


def euclidean_dist(a, b):
    d=0
    for i, j in zip(a, b):
        d+=(i-j)**2
    return d**0.5
    
```

## Â¿QuÃ© parÃ¡metro arroja mÃ¡s verosimilitud?

```python
diccionario={}
```

```python
for lamda in range(100):
    print(lamda, end="\r")
    lam=lamda/50.0
    for row in abundance_t.iterrows():
        fams=dict(row[1])
        if row[0] not in diccionario: diccionario[row[0]]={}
        if lam not in diccionario[row[0]]:
            xs=[]
            ys=[]
            for familia in fams:
                if fams[familia]==1: 
                    xs.append(information.loc[familia]["X"])
                    ys.append(information.loc[familia]["Y"])
            if len(xs)>1:
                buenas=len(xs)
                malas=len(fams)-buenas
                loc=(np.mean(xs), np.mean(ys))

                fam_ping=1
                for familia in fams:
                    distancia=euclidean_dist(loc, (information.loc[familia]["X"], information.loc[familia]["Y"]))
                    prob=exponential_distribution(lam, distancia)
                    if fams[familia]==1:
                        fam_ping=fam_ping*(prob**(malas/buenas))
                    else:
                        fam_ping=fam_ping*(1-prob)

                diccionario[row[0]][lam]=(fam_ping**(1/(2*malas)))
            else:
                diccionario[row[0]][lam]=1
            
```

```python
verosimilitudes=[]
lams=[]
for lamda in range(1,100):
    lam=lamda/50.0
    verosimilitud=1
    for row in abundance_t.iterrows():
        verosimilitud=verosimilitud*(diccionario[row[0]][lam]*10)
    verosimilitudes.append(verosimilitud)
    lams.append(lam)
```

```python
plt.plot(lams, verosimilitudes)
plt.xlim([0.08,0.25])
```
