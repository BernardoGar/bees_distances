```python
def get_htags(texto):
    ts=texto.replace("\n", " ").split("#")
    hashtags=[]
    i=1
    while i<len(ts):
        fin=ts[i].find(" ")
        if fin>0:
            hashtags.append(ts[i][:fin])
        else:
            hashtags.append(ts[i])
        i+=1
    return hashtags
```

```python
def get_sim_texts(t1, t2, distr, p_so):
    #p_t1_equal_t1=prob_that(t1, distr)
    p_t2_equal_t2_given_t1=prob_that_given(t2, distr, t1)
    p_t2_equal_t2=prob_that(t2, distr)
    #num=p_t1_equal_t1*p_t2_equal_t2_given_t1*p_so
    #denom=num+(1-p_so)*p_t2_equal_t2*p_t1_equal_t1
    num=p_t2_equal_t2_given_t1*p_so
    denom=num+(1-p_so)*p_t2_equal_t2
    p_so_given_t1_t2=num/denom
    return p_so_given_t1_t2


def prob_that(t1, distr):
    a=1
    j=0
    for mot in t1.split(" "):
        a=a*distr[mot]
        j+=1
    return a**(1/j)

def prob_that_given(t2, distr, t1):
    a=1
    j=0
    for mot in t2.split(" "):
        new_prob=0.5*distr[mot]+int(mot in t1.split(" "))*0.5
        a=a*new_prob
        j+=1
    return a**(1/j)
```

```python
alldfs2={}
#ids_bernie=[949]
for idb in ids_bernie:
    print(idb)
    connection=get_connection()

    with connection.cursor() as cursor:
        sql="select * from tweets_"+str(idb)
        cursor.execute(sql)
        tweets=cursor.fetchall()

        sql="select * from users_"+str(idb)
        cursor.execute(sql)
        users=cursor.fetchall()

    connection.close()

    tweets=pd.DataFrame(tweets)
    users=pd.DataFrame(users)

    all_ids=list(set(users["user"]))



    all_ids=tweets["user"]
    all_htags={}
    j=0

    for id1 in all_ids:
        j+=1
        print(str(j)+"/"+str(len(all_ids)), end="          \r")
        hashtags_user=[]
        for testo in users[users.user==id1]["texto"]:
            hashtags_user+=get_htags(testo.lower())
        hashtags_user=list(set(hashtags_user))
        all_htags[id1]=hashtags_user

    all_htags_list=[]
    for id1 in all_ids:
        all_htags_list+=all_htags[id1]

    all_htags_list=list(set(all_htags_list))
    count={}
    for ht in all_htags_list:
        count[ht]=0
    for id1 in all_ids:
        for ht in all_htags[id1]:
            count[ht]+=1


    numbers=[count[ht] for ht in count]

    # the histogram of the data
    n, bins, patches = plt.hist(numbers, 20, density=True, facecolor='g', alpha=0.75)


    plt.xlabel('Smarts')
    plt.ylabel('Probability')
    plt.title('Histogram of occurenced of HT')
    plt.ylim(0,0.0001)
    plt.grid(True)
    plt.show()

    #Better to use complete jaccard distance so it is interpretable in an absolute way

    simis=[]
    sim_matrix=np.zeros([len(all_ids), len(all_ids)])


    j=0
    for id1 in all_ids:
        print(str(j)+"/"+str(len(all_ids)), end="         \r")
        i=0

        with terminating(Pool(processes=4)) as pool:
            results = pool.map(partial(get_row_jac, j=j, all_hashtags=all_htags, all_ids=all_ids), range(len(all_ids)))

        for id2 in all_ids:
            if i==j:
                simis.append(1)
                sim_matrix[i,j]=1
            if i>j:
                ssim=results[i]
                simis.append(ssim)
                sim_matrix[j,i]=ssim
                sim_matrix[i,j]=ssim
            i+=1

        j+=1
        
    
    textos=[]
    for tweet in tweets["texto"]:
        textos.append(tweet.lower().replace("\n", " "))
        
        
      
    mots=[]
    for texto in textos:
        mots+=list(set(texto.split(" ")))
    mots_all=list(set(mots))
    distr={}
    for mot in mots_all:
        distr[mot]=0

    sumi=1/len(textos)
    for mot in mots:
        distr[mot]+=sumi

        
    sims_gen=np.zeros((len(textos), len(textos)))
    p_so=700/5000


    j=0
    for texto1 in textos:
        print(str(j)+"/"+str(len(textos)), end="          \r")
        i=0
        for texto2 in textos:
            if i==j: 
                sim=1
                sims_gen[i][j]=sim
            if i>j:
                sim=get_sim_texts(texto1, texto2, distr, p_so)
                sims_gen[i][j]=sim

            i+=1
        j+=1

    j=0
    
    nl=[]
    j=0
    all_list=[]
    for texto1, u1 in zip(textos, all_ids):
        print(str(j)+"/"+str(len(tweets)), end="          \r")
        i=0
        for texto2, u2 in zip(textos, all_ids):
            if u1>u2 and (u1, u2) not in all_list and np.min([len(set(all_htags[u1])), len(set(all_htags[u1]))])>20 :
                nd={"sim":sims_gen[i][j], "sim_user":sim_matrix[i][j], "t1":texto1, "t2":texto2, "u1":u1, "u2":u2}
                nl.append(nd)
            i+=1
        j+=1

    j=0

    df2=pd.DataFrame(nl)
    
    alldfs2[idb]=df2
    
    
```
