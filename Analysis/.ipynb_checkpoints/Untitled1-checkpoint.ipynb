```python
import random
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
math.cos(1*math.pi)
```

```python
def euclidean(a, b):
    dis=0
    for aa, bb in zip(a, b):
        dis+=(aa-bb)**2
    return dis**0.5

def p_visit(d):
    proba=(1-abs(d-0.1))*0.05
    if proba<0:
        proba=0
    return proba

def visited_hives(origin, beehives):
    i=0
    for k in beehives:
        pvisit=p_visit(euclidean(origin, k))
        if random.random()<pvisit:
            return i
        i+=1
    return -1
    
    
beehives=[]
for k in range(30):
    beehives.append((np.random.normal(0,20), np.random.normal(0,20)))

visits=[]
travels=[]
for bee in range(1000):
    print(bee, end="        \r")
    step=0
    place=(0,0)
    dtravelled=0
    while step<10000:
        flight=random.random()/3
        angle=2*random.random()*math.pi
        place=(place[0]+math.sin(angle)*flight, place[1]+math.cos(angle)*flight)
        vis=visited_hives(place, beehives)
        if vis>-1: step=100000+vis
        else: step+=1
    dtravelled=euclidean(place, (0,0))
    if step>20000:
        visit=step-100000
        visits.append(visit)
    travels.append(dtravelled)
```

```python
np.mean(travels)
```

```python
def visited_hives2(origin, beehives, thresh):
    i=0
    l=list(range(len(beehives)))
    random.shuffle(l)
    for k in l:
        pvisit=euclidean(origin, beehives[k])
        if pvisit<thresh:
            return k
    return -1
```

```python
distances=[]
tots=[]
for k in range(len(beehives)):
    bep=beehives[k]
    tot=0
    for p in visits:
        if p==k: tot+=1
    distances.append(euclidean(bep, (0,0)))
    tots.append(tot)
```

```python

visits2=[]
for bee in range(1000):
    print(bee, end="        \r")
    step=0
    place=(0,0)
    flight=np.random.exponential(8)
    angle=2*random.random()*math.pi
    place=(place[0]+math.sin(angle)*flight, place[1]+math.cos(angle)*flight)
    #print(place)
    vis=visited_hives2(place, beehives,  7)
    if vis>-1: 
        visits2.append(vis)

tots2=[]
for k in range(len(beehives)):
    tot=0
    for p in visits2:
        if p==k: tot+=1
    tots2.append(tot)
```

```python
np.sum(tots2)
```

```python
plt.scatter(distances, tots, label="Random walk (100m catchment area)")
plt.scatter(distances, tots2, label="Exponential arrival destination fit (7km catchment area)")
plt.legend()
plt.yscale("log")
plt.ylim([0.9, 500])
plt.xlabel("Km of beehive to origin point")
plt.ylabel("# of bees captured")
plt.savefig("example.pdf")
```
