```python
import matplotlib

matplotlib.rc('font',**{'family':'serif','serif':['Helvetica'], 'size':11})
plt.rcParams['pdf.fonttype'] = 42


plt.figure(figsize=(8,5))

mean=0.5
for mean in  [5.0, 2.5]:
    plt.plot([i/1000 for i in range(50000)],[1-math.exp((-(i/1000)/mean)) for i in range(50000)], label="Mean: "+str(mean))
#plt.title("Cumulative distribution of exponential variables")
plt.xlim([0,30])
plt.legend(bbox_to_anchor=(1,1))
plt.ylabel("Probability function of exponential distribution")
plt.xlabel("Flown distance (km)")
plt.savefig("exponential_cum.pdf")
plt.show()







plt.figure(figsize=(14,14))
mean=0.5
for mean in [ 5.0, 2.5]:
    plt.plot([i/1000 for i in range(50000)],[(1/mean)*math.exp((-(i/1000)/mean)) for i in range(50000)], label="Mean: "+str(mean))
plt.title("Distribution of exponential variables")
plt.legend(bbox_to_anchor=(1,1), prop={"size":30})
plt.ylim([0,0.4])
plt.xlim([0,15])
plt.ylabel("Probability of male dispersal", size=30)
plt.xticks(size=30)
plt.yticks([i/10.0 for i in range(5)], size=30)
plt.xlabel("Distance male disperses (km)", size=30)
plt.savefig("exponential812.pdf", bbox_inches="tight")
plt.show()

```
