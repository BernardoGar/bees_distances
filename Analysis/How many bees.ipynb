```python
import pandas as pd
import numpy as np
```

```python
f=open("output_colony.csv", "r")
a="A"
ind=0
familias=[]
while a!="":
    a=f.readline()
    if ind>0:
        nd={}
        b=a.split(",")
        if len(b)>2:
            nd["Familia"]=b[0]
            nd["c1"]=float(b[1])
            nd["c2"]=float(b[2])
            ind2=3
            parientes=[]
            while ind2<len(b):
                if b[ind2]!="" and b[ind2]!="\n": parientes.append(b[ind2].replace("\n", ""))
                ind2+=1
            nd["integrantes"]=parientes
            familias.append(nd)
    ind+=1
    
    
```

```python
df=pd.read_csv("Aleles_clean.csv")
df_syd2018=df[df.Cluster=="PC2018"].copy()
```

```python
dict_fam={}
for f in familias:
    fam=f["Familia"]
    for i in f["integrantes"]:
        dict_fam[i]=fam
```

```python
df_syd2018["family"]=[dict_fam[i] for i in df_syd2018["name1"]]
```

```python
ress=[]
for p in range(1,2118):
    nl=[]
    print(p, end="      \r")
    for s in range(100):
        nl.append(len(set(df_syd2018.sample(p)["family"])))
    ress.append(nl)
    
```

```python
import matplotlib.pyplot as plt
```

```python
dict_save_all={}
dict_save_all["Syd2018"]=([np.percentile(i, 5) for i in ress], [np.percentile(i, 95) for i in ress],[np.mean(i) for i in ress] )

```

```python
plt.fill_between(range(1,2118), [np.percentile(i, 5) for i in ress], [np.percentile(i, 95) for i in ress], color="gray")
plt.plot(range(1,2118),  [np.mean(i) for i in ress])
plt.xlabel("Sampled bees")
plt.ylabel("Families represented\n(According to COLONY)")
plt.savefig("HowmanySyd2018.pdf")
```

```python
tot_fam=len(list(set(df_syd2018["family"])))
flag=0
i=1
for s in [np.mean(i) for i in ress]:
    if s/tot_fam>=0.8 and flag==0:
        flag=1
        print(i)
    i+=1
```

## QL by site

```python
dict_save_all_ql={}
```

```python
for farm in ["saharafarms", "mooloolah", "beerwah"]:
    f=open("/Users/Bernardo/Dropbox/abejas_temp/Output_colony_"+farm+".csv", "r")
    a="A"
    ind=0
    familias=[]
    while a!="":
        a=f.readline()
        if ind>0:
            nd={}
            b=a.split(",")
            if len(b)>2 and len(str(b[0]))>0:
                nd["Familia"]=b[0]
                nd["c1"]=float(b[1])
                nd["c2"]=float(b[2])
                ind2=3
                parientes=[]
                while ind2<len(b):
                    if b[ind2]!="" and b[ind2]!="\n": parientes.append(b[ind2].replace("\n", ""))
                    ind2+=1
                nd["integrantes"]=parientes
                familias.append(nd)
        ind+=1

    dict_fam={}
    for f in familias:
        fam=f["Familia"]
        for i in f["integrantes"]:
            dict_fam[i]=fam



    nl=[]
    for fam in familias:
        for f in fam["integrantes"]:
            nl.append({"Integrante":f, "Familia":fam["Familia"]})
    df=pd.DataFrame(nl)
    ress=[]
    totfams=len(set(df["Familia"]))
    for p in range(1,len(df)):
        nl=[]
        print(p, end="      \r")
        for s in range(100):
            nl.append(len(set(df.sample(p)["Familia"])))
        ress.append(nl)
    dict_save_all_ql[farm]=([np.percentile(i, 5) for i in ress], [np.percentile(i, 95) for i in ress],[np.mean(i) for i in ress] )
    plt.fill_between(range(1,len(df)), [np.percentile(i, 5) for i in ress], [np.percentile(i, 95) for i in ress], color="gray")
    plt.plot(range(1,len(df)),  [np.mean(i) for i in ress])
    for k in [.8, .85, .9, .95]:
        ii=1
        while [np.mean(i)/totfams for i in ress][ii] < k : ii+=1
        print("To reach "+str(k)+" we need "+str(ii)+" bees out of  "+str(len(df))+" bees in "+farm)
            
            
            
        
    plt.xlabel("Sampled bees")
    plt.ylabel("Families represented\n(According to COLONY)")
    plt.savefig("HowmanyQL"+farm+".pdf")
    plt.show()

```
